Leitura da base de dados

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

vendas_df = pd.read_csv("Dataset.csv")
vendas_df
```

Retorna valores nulos

```python
vendas_df[vendas_df.isna().any(axis=1)]
```

Remove valores nulos

```python
vendas_df = vendas_df.dropna()
```

Remove duplicatas

```python
vendas_df = vendas_df.drop_duplicates(subset=['order_id'])
vendas_df
```

Retorno de informações gerais como:

    Receita total
    Lucro total
    Ticket médio
    Quantidade total vendida
    Margem de lucro (%)

```python
description_sale = {
    'Receita_Total': None,
    'Lucro_Total': None,
    'Ticket_Medio': None,
    'QTD_Vendida': None,
    'Margem_Lucro': None
}

description_sale['Receita_Total'] = vendas_df['total_value'].sum()
description_sale['Lucro_Total'] = vendas_df['profit'].sum()
description_sale['Ticket_Medio'] = vendas_df['total_value'].mean()
description_sale['QTD_Vendida'] = vendas_df['quantity'].sum()
mg_result = (description_sale['Receita_Total'] - description_sale['Lucro_Total']) * 100
description_sale['Margem_Lucro'] = mg_result / description_sale['Receita_Total'] *100
print(description_sale['Margem_Lucro'])

```
