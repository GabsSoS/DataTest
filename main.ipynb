Leitura da base de dados

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

vendas_df = pd.read_csv('vendas.csv')
vendas_df['Custo_Final'] = vendas_df['Preco_Custo'] * vendas_df['Quantidade']
vendas_df.head(5)
```

Métricas Principais
    
	Receita de Venda Total
	Lucro Total
	Ticket médio
	Quantidade de Venda Total
	Margem de lucro (%)

```python
metricas_principais = {
    'Receita_de_Venda_Total':None,
    'Lucro_Total':None,
    'QTD_Vendas_Total':None,
    'Ticket_Medio': None,
    'Margem_de_Lucro':None,
}

metricas_principais['Receita_de_Venda_Total'] = vendas_df['Valor_Final'].sum()
metricas_principais['Lucro_Total'] = vendas_df['Valor_Final'].sum() - vendas_df['Custo_Final'].sum()
metricas_principais['QTD_Vendas_Total'] = vendas_df['Quantidade'].sum()
metricas_principais['Ticket_Medio'] = metricas_principais['Receita_de_Venda_Total'] / metricas_principais['QTD_Vendas_Total']
metricas_principais['Margem_de_Lucro'] = metricas_principais['Lucro_Total'] / metricas_principais['Receita_de_Venda_Total'] * 100

```
